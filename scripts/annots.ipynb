Converts 6 classes generated by the roboflow to the 5 original classes (removes the 'object' class

```python
path = '../data/rsdcoco/annotations/instances_val.json'
save_path = '../data/rsdcoco/annotations/instances_val_copy.json'

mapping = {
    1:0,
    2:1,
    3:2,
    4:3,
    5:4
}

import json
# Opening JSON file
f = open(path)

# returns JSON object as
# a dictionary
data = json.load(f)
#data_copy = data.copy()

#Remove the object class:
for i,d in enumerate(data['categories']):
    if d['name'] != 'objects' and d['name'] != 'object':
        data['categories'][i]['id'] = mapping[d['id']]

data['categories'].pop(0)


# Iterating through the json list to remap the Roboflow annotations
for i,d in enumerate(data['annotations']):
    #print(d)
    data['annotations'][i]['category_id'] = mapping[d['category_id']]


# Closing file
f.close()

# Save to the new file
with open(save_path, 'w') as f:
    json.dump(data, f, indent=4)
```
